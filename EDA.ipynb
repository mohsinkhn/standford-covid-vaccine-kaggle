```python
class RNAGRUModel(nn.Module):
    def __init__(self, hparams):
        super().__init__()
        self.hparams = hparams
        self.init_params()
        self.embedding = nn.Embedding(len(token2int.keys()), self.emb_dim)
        self.gru = nn.GRU(input_size=self.emb_dim * self.num_features, hidden_size=self.gru_dim,
                          bidirectional=self.bidirectional, batch_first=True,
                         )
        self.fc = nn.Sequential(
            nn.Linear(self.gru_dim * (1 + self.bidirectional), self.target_dim)
        )

    def init_params(self):
        self.emb_dim = self.hparams.get("emb_dim", 64)
        self.gru_dim = self.hparams.get("gru_dim", 32)
        self.bidirectional = self.hparams.get("bidirectional", True)
        self.target_dim = self.hparams.get("target_dim", 5)
        self.num_features = self.hparams.get("num_features", 3)
        self.max_seq_pred = self.hparams.get("max_seq_pred", 68)

    def forward(self, x):
        x = self.embedding(x)
        bs, seq_len = x.size()[:2]
        x = x.contiguous().view(bs, seq_len, -1)
        x, _ = self.gru(x)
        x = x[:, :self.max_seq_pred, :]
        x = self.fc(x)
        return x
    
    
```

```python
class MCRMSE(nn.Module):
    def __init__(self):
        super().__init__()
    
    def forward(self, outputs, targets):
        sqaured_error = (outputs - targets)**2
        root_mean_sqaured_error = torch.sqrt(torch.mean(sqaured_error, dim=(0, 2)))
        return root_mean_sqaured_error.mean()
```
